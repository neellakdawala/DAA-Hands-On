## Problem -1 

Leverage your implementation of quicksort to implement the ith order statistic. Demonstrate it's working via an example. Upload your code to github.




```python
def partition(arr, low, high):
    pivot = arr[high]
    i = low - 1
    
    for j in range(low, high):
        if arr[j] <= pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]
    
    arr[i + 1], arr[high] = arr[high], arr[i + 1]
    return i + 1

def quickselect(arr, low, high, i):
    if low == high:
        return arr[low]
    
    pivot_index = partition(arr, low, high)
    
    k = pivot_index - low + 1
    
    if i == k:
        return arr[pivot_index]
    elif i < k: 
        return quickselect(arr, low, pivot_index - 1, i)
    else:
        return quickselect(arr, pivot_index + 1, high, i - k)

arr = [12, 3, 5, 7, 19, 26, 2, 9]
i = 5  # Find the 5th smallest element
print(f"{i}th smallest element is {quickselect(arr, 0, len(arr) - 1, i)}")
```

```python
class Stack:
    def __init__(self, size):
        self.stack = [0] * size  
        self.top = -1            
        self.size = size         

    def is_empty(self):
        return self.top == -1

    def is_full(self):
        return self.top == self.size - 1

    def push(self, value):
        if self.is_full():
            print("Stack overflow")
            return
        self.top += 1
        self.stack[self.top] = value
        print(f"Pushed {value} onto the stack.")

    def pop(self):
        if self.is_empty():
            print("Stack underflow")
            return None
        value = self.stack[self.top]
        self.top -= 1
        print(f"Popped {value} from the stack.")
        return value

    def peek(self):
        if self.is_empty():
            print("Stack is empty.")
            return None
        return self.stack[self.top]

    def display(self):
        if self.is_empty():
            print("Stack is empty.")
            return
        print("Stack content from top to bottom:")
        for i in range(self.top, -1, -1):
            print(self.stack[i])

stack = Stack(5)
stack.push(10)  # Pushes 10
stack.push(20)  # Pushes 20
stack.push(30)  # Pushes 30
stack.push(40)  # Pushes 40
stack.push(50)  # Pushes 50
stack.push(60)  # Pushes 60
stack.display()
print(f"Top element is: {stack.peek()}")
stack.pop()  # Pops 50
stack.pop()  # Pops 40
stack.display()
```

```python
class Queue:
    def __init__(self, size):
        self.queue = [0] * size  
        self.front = 0           
        self.rear = -1          
        self.size = size         
        self.count = 0       

    def is_empty(self):
        return self.count == 0

    def is_full(self):
        return self.count == self.size

    def enqueue(self, value):
        if self.is_full():
            print("Queue overflow")
            return
        self.rear = (self.rear + 1) % self.size
        self.queue[self.rear] = value
        self.count += 1
        print(f"Enqueued {value} into the queue.")

    def dequeue(self):
        if self.is_empty():
            print("Queue underflow")
            return None
        value = self.queue[self.front]
        self.front = (self.front + 1) % self.size
        self.count -= 1
        print(f"Dequeued {value} from the queue.")
        return value

    def peek(self):
        if self.is_empty():
            print("Queue is empty.")
            return None
        return self.queue[self.front]

    def display(self):
        if self.is_empty():
            print("Queue is empty.")
            return
        print("Queue content from front to rear:")
        idx = self.front
        for i in range(self.count):
            print(self.queue[idx], end=" ")
            idx = (idx + 1) % self.size
        print()

queue = Queue(5)
queue.enqueue(10)  # Enqueues 10
queue.enqueue(20)  # Enqueues 20
queue.enqueue(30)  # Enqueues 30
queue.enqueue(40)  # Enqueues 40
queue.enqueue(50)  # Enqueues 50
queue.enqueue(60)  # Enqueues 60
queue.display()   
print(f"Front element is: {queue.peek()}") 
queue.dequeue()    # Dequeues 10
queue.display()
```

```python
class SinglyLinkedList:
    def __init__(self, size):
        self.list = [None] * size  
        self.next = [-1] * size    
        self.head = -1             
        self.free = 0             
        self.size = size
        
        for i in range(size - 1):
            self.next[i] = i + 1
        self.next[size - 1] = -1

    def insert(self, value):
        if self.free == -1:
            print("List overflow")
            return
        
        new_node = self.free
        self.free = self.next[self.free]
        self.list[new_node] = value
        self.next[new_node] = self.head
        self.head = new_node

    def delete(self, value):
        if self.head == -1:
            print("List underflow")
            return
        
        prev, current = -1, self.head
        while current != -1 and self.list[current] != value:
            prev = current
            current = self.next[current]
        
        if current == -1:
            print(f"{value} not found")
            return
        
        if prev == -1:
            self.head = self.next[current]
        else:
            self.next[prev] = self.next[current]
        
        self.next[current] = self.free
        self.free = current

    def display(self):
        current = self.head
        while current != -1:
            print(self.list[current], end=" -> ")
            current = self.next[current]
        print("None")

ll = SinglyLinkedList(5)
ll.insert(10)
ll.insert(20)
ll.display() 
ll.delete(10)
ll.display()
```
